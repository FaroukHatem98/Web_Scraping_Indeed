```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
driver = webdriver.Firefox()
```

```python
driver.get('https://eg.indeed.com/')
```

```python
WebDriverWait(driver, 50).until(ec.presence_of_element_located((By.XPATH, '//*[@id="text-input-what"]')))
```

```python
search = driver.find_element("xpath",'//*[@id="text-input-what"]')
search.send_keys('data analyst')
search.send_keys(Keys.ENTER)
```

```python
df = pd.DataFrame({'job':[''],'company name':[''], 'date':[''], 'job type':[''], 'salary':[''], 'link':['']}, index = ['job'])
df = df.head(0)
```

```python
time.sleep(5)
```

```python
while True:
    time.sleep(2)
    soup = bs(driver.page_source, 'lxml')
    posts = soup.find_all('div', class_ = 'slider_container css-g7s71f eu4oa1w0')
    for post in posts:
        job = post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').text
        link = 'https://eg.indeed.com' + post.find('a', class_ = 'jcs-JobTitle css-jspxzf eu4oa1w0').get('href')
        date = post.find('span', class_ = 'date').text
        if 'Posted' in date:
            date = date[6:]
        if 'Posted' in date:
            date = date[6:]
        try :
            companyname = post.find('span', class_ = 'companyName').text
        except:
            companyname = 'unknown'
        s = post.find_all('div', class_ = 'attribute_snippet')
        if len(s) == 2:
            job_type = s[1].text
            Salary = s[0].text
        elif len(s) == 1:
            job_type = s[0].text
            Salary = 'unknown'
        else :
            jop_type = 'unknown'
            Salary = 'unknown'
        row = [job, companyname, date, job_type, Salary, link]
        lent = len(df)
        df.loc[lent] = row
    try :
        try:
            next_page = soup.find('a', {'aria-label' : 'Next Page'}).get('href')
        except:
            next_page = soup.find('a', {'aria-label' : 'Next'}).get('href')
    except:
        break
    next_page = 'https://eg.indeed.com' + next_page
    driver.get(next_page)
```

```python
df.to_csv('indeed.csv', index= False)
```
